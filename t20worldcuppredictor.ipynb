```python
import numpy as np
import pandas as pd
```

```python
mdf = pd.read_csv('matches.csv')
ddf = pd.read_csv('deliveries.csv')
```

```python
mdf.head()
```

```python
ddf.head()
```

```python
inning_score_df = ddf.groupby(['match_id','innings']).sum()['total_runs'].reset_index()
inning_score_df = inning_score_df[inning_score_df['innings'] == 1]
inning_score_df
```

```python
mdf['id']
```

```python
mdff = mdf.merge(inning_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
mdff
```

```python
mdff = mdff[['match_id','venue','winner','total_runs']]
ddff = mdff.merge(ddf,on='match_id')
ddff = ddff[ddff['innings'] == 2]
ddff
```

```python
ddff['current_score'] = ddff.groupby('match_id').cumsum()['total_runs_y']
ddff['runs_left'] = ddff['total_runs_x'] - ddff['current_score'] + 1
ddff['balls_left'] = 126 - (ddff['over']*6 + ddff['balls'])
ddff
```

```python
ddff['player_dismissed'] = ddff['player_dismissed'].fillna("0")
ddff['player_dismissed'] = ddff['player_dismissed'].apply(lambda x:x if x == "0" else "1")
ddff['player_dismissed'] = ddff['player_dismissed'].astype('int')
wickets = ddff.groupby('match_id').cumsum()['player_dismissed'].values
ddff['wickets'] = 10 - wickets
ddff.head()
```

```python
ddff['crr'] = (ddff['current_score']*6)/(120 - ddff['balls_left'])
```

```python
# rrr = runs_left/overs_left
ddff['rrr'] = (ddff['runs_left']*6)/ddff['balls_left']
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
ddff['result'] = ddff.apply(result,axis=1)
```

```python
finalDf = ddff[['batting_team','bowling_team','venue_x','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
finalDf = finalDf.sample(finalDf.shape[0])
```

```python
finalDf.sample()
```

```python
finalDf.dropna(inplace=True)
```

```python
finalDf = finalDf[finalDf['balls_left'] != 0]
X = finalDf.iloc[:,:-1]
y = finalDf.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','venue_x'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.pipeline import Pipeline
import pickle
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2', AdaBoostClassifier())
])
```

```python
pipe.fit(X_train,y_train)
```

```python
filename = 'LR.sav'
pickle.dump(pipe, open(filename, 'wb'))
```

```python
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(X_test, y_test)
print(result)
```

```python
y_pred = pipe.predict(X_test)
```

```python
pipe.predict_proba(X_test)[10]
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print('AdaBoostClassifier')
print(classification_report(y_test, y_pred))
print('confusion_matrix')
print(confusion_matrix(y_test, y_pred))
```
